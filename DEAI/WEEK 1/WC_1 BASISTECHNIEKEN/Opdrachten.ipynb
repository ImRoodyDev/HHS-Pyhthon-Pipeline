## While-loop

Schrijf de inhoud van de methode wachtwoord. Deze heeft geen parameters en laat de gebruiker twee wachtwoorden invoeren, waarna hij bekijkt of deze gelijk zijn. Zolang dat niet het geval is wordt er eerst Fout, probeer het opnieuw! uitgeprint en moet de gebruiker het opnieuw proberen.<br>Als de gebruiker het goed doet, komt er Succes! op het scherm te staan.

```python
def wachtwoord() :
    password = input("Voer wachtwoord in:")
    passed: bool= False
    while not passed: 
        if password == input("Bevestig wachtwoord:"):
            print("Wachtwoord geaccepteerd")
            passed = True
        else: 
            print("Wachtwoorden komen niet overeen, probeer opnieuw.")
            
wachtwoord()
```

## Objectoriëntatie

Maak een minisysteem voor een kledingwinkel. Maak verschillende klassen voor verschillende soorten kledingstukken, waarbij sommige klassen geërfd worden van andere klassen.
- Maak een klasse Kledingstuk aan, waarin door de constructor een merk, een maat en een prijs geïnitialiseerd worden.
- Maak een klasse Broek en een klasse Shirt, beide erven van de klasse Kledingstuk. Voeg aan Broek een broek_type, en aan Shirt een shirt_type toe (d.m.v. constructorgebruik).
- Maak een klasse Spijkerbroek die erft van broek en een klasse Poloshirt die erft van shirt. Voeg aan Spijkerbroek een blauwtint toe, en aan Poloshirt een kraagsoort (d.m.v. constructorgebruik).
- Implementeer in elke klasse een beschrijvingmethode, waarmee alle informatie over het betreffende kledingstuk wordt afgedrukt.
- Maak van elke klasse een object aan en roep de beschrijvingmethode aan. Controleer nauwkeurig of alles naar behoren werkt.

```python
from dataclasses import dataclass

@dataclass
class Kledingstuk:
    merk: str
    maat: str
    prijs: float
    
    def info(self):
      return self.__repr__()
        
@dataclass
class Broek(Kledingstuk):
    broek_type: str
 
@dataclass    
class Shirt(Kledingstuk):
    shirt_type: str
  
class SpijkerBroek(Broek):
    def __init__(self, blauwtint, broek_type, merk, maat, prijs):
        super().__init__(merk, maat, prijs, broek_type)
        self.blauwtint = blauwtint
        
    def info(self):
        return(f"Spijkerbroek van het merk {self.merk}, maat {self.maat}, prijs {self.prijs:.2f}, broek type {self.broek_type} en blauwtint {self.blauwtint}")
        
class PoloShirt(Shirt):
    def __init__(self, kleur, shirt_type, merk, maat, prijs):
        super().__init__(merk, maat, prijs, shirt_type)
        self.kleur = kleur
        
    def info(self):
        return(f"Polo shirt van het merk {self.merk}, maat {self.maat}, prijs {self.prijs:.2f}, shirt type {self.shirt_type} en kleur {self.kleur}")
    
# --- Create objects ---
k  = Kledingstuk("Zara", "M", 19.99)
b  = Broek("Nike", "M", 49.99, "Chino")
s  = Shirt("Adidas", "L", 29.99, "T-shirt")
sb = SpijkerBroek("Lichtblauw", "Slim fit", "Levi's", "32", 79.99)
ps = PoloShirt("Wit", "Polo", "Ralph Lauren", "XL", 89.99)

# --- Call beschrijving ---
print(k.info())
print(b.info())
print(s.info())
print(sb.info())
print(ps.info())
```
